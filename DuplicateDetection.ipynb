```python
# Inporting the Libraries
import os
import shutil
from PIL import Image
import imagehash

#Directory Paths for Input & Output - (Duplicate & Not Duplicate)
dirname = "  ---  Input Path  ---  "
duplicate_dir = "  ---  Output Path for DUPLICATE IMAGES  ---  "
unique_dir = "  ---  Output Path for NOT DUPLICATE IMAGES  ---  "

total_images = len(os.listdir(dirname))
print("Threshold - 5")
print("Total Images: ", total_images)

#Converting all images into array
image_names = os.listdir(dirname)
hashes = {}

#Empty List for Duplicate Images
duplicate_image_list = list()

#Empty List for Not Duplicate Images
unique_image_list = list()

#Threshold value as per requirements
hash_size = 5

# FOR Loop - To iterate on all the images of array
for image in image_names:
    with Image.open(os.path.join(dirname, image)) as img:
        temp_hash = imagehash.average_hash(img, hash_size)

#Iterations on Duplicate Images
        if temp_hash in hashes:
            duplicate_image_list.append(image)
        else:

#Iterations on Not Duplicate (Unique) Images
            hashes[temp_hash] = image
            unique_image_list.append(image)

for image in duplicate_image_list:
    Image_Full_Path = str(dirname) + image

#To COPY the Duplicate Images to the Duplicate Image folder as per the IF Condition
    shutil.copy(Image_Full_Path, duplicate_dir)

for image in unique_image_list:
    uImage_Full_Path = str(dirname) + image
    
#To COPY the Not Duplicate Images to the Not Duplicate Image folder as per the IF Condition
    shutil.copy(uImage_Full_Path, unique_dir)
        
#To Print the Count of Total Duplicate Images from Duplicate Image Folder    
print("Duplicate Images: ", len(os.listdir(duplicate_dir)))

#To Print the Count of Total Not Duplicate Images from Not Duplicate Image Folder    
print("Unique Images: ", len(os.listdir(unique_dir)))
```
